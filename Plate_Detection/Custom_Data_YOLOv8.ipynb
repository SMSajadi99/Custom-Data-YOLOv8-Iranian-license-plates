## Convert Convert .avi to .mp4

```
!pip install moviepy
```

```
from moviepy.editor import VideoFileClip

def avi_to_mp4(input_file, output_file):
    try:
        # Load the AVI file
        clip = VideoFileClip(input_file)

        # Set the output file format to mp4 and save the video
        clip.write_videofile(output_file, codec='libx264')

        print(f"Conversion completed successfully! Saved as '{output_file}'")
    except Exception as e:
        print(f"Error during conversion: {e}")

if __name__ == "__main__":
    input_filename = r"/content/runs/detect/predict2/Car.avi"    # Replace with the name of your input .avi file
    output_filename = r"/content/runs/detect/predict2/Car_mp4.mp4"  # Replace with the desired name for the output .mp4 file

    avi_to_mp4(input_filename, output_filename)
```
